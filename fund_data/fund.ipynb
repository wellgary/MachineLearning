# 將500檔基金的日期合併(以日期最多的基金為基準)

### 先找出日期最多的基金

```python
max_len = len(group_fund_data.get_group(unique_code[0]))
max_len_index = 0
for i in range(1,500):
    loop_len = len(group_fund_data.get_group(unique_code[i]))
    if(loop_len > max_len):
        max_len = loop_len
        max_len_index = i
```

```python
max_len
```

```python
max_len_index
```

### 將第93檔基金放到unique_code的第一個，依序將其他放入

```python
revise_unique_code = []
revise_unique_code.append(unique_code[92])
```

```python
for i in range(0,92):
    revise_unique_code.append(unique_code[i])
for i in range(93,500):
    revise_unique_code.append(unique_code[i])
```

```python
revise_unique_code[0:5]
```

### 以第93檔基金的日期做為基準，並合併

```python
outside_data = group_fund_data.get_group(revise_unique_code[0])
outside_data.index = outside_data.date
outside_data = outside_data.iloc[:,2:3]
outside_data.columns = [revise_unique_code[0]]
```

```python
outside_data.head()
```

```python
for i in range(1, len(unique_code)):
    loop_data = group_fund_data.get_group(unique_code[i])
    loop_data.index = loop_data.date
    loop_data = loop_data.iloc[:,2:3]
    loop_data.columns = [unique_code[i]]
    outside_data = pd.concat([outside_data, loop_data], axis=1, join_axes=[outside_data.index])
```

```python
merge_fund_data = outside_data.dropna()
```

```python
merge_fund_data.head()
```

```python
merge_fund_data["date"] = merge_fund_data.index
```

```python
merge_fund_data.index = range(0,len(merge_fund_data))
```

```python
merge_fund_data.head()
```

```python
for i in range(0, 500):
    loop_data = merge_fund_data.iloc[:,i:i+1]
    for j in range(0, int(len(merge_fund_data)/5)):
        code = revise_unique_code[i]
        start_date = merge_fund_data["date"][5*j]
        end_date = merge_fund_data["date"][5*j+4]
        OPEN = loop_data.iloc[j*5, 0]
        CLOSE = loop_data.iloc[j*5+4, 0]
        HIGH = max(loop_data.iloc[j*5:(j*5+5), 0])
        LOW = min(loop_data.iloc[j*5:(j*5+5), 0])
        dd = {'code':[code], 'start':[start_date], 'end':[end_date], 'open':[OPEN], 'close':[CLOSE], 'high':[HIGH], 'low':[LOW]}
        ddf = pd.DataFrame(data=dd)
        df = pd.concat([df,ddf],axis=0)
    print(i)
```
